```
 %tensorflow_version 2.x
import tensorflow as tf
import numpy as np
import pandas as pd
import keras.backend as K
from tensorflow import keras
from keras.optimizers import SGD
from tensorflow.python.keras.layers import Dense
from sklearn.model_selection import train_test_split 
```

```
data = pd.read_csv("computer_hardware_weka_dataset.csv")
data.head()
data.drop(['vendor_name','model_name'], axis=1, inplace=True)
```

```
X = data.iloc[:,0:7]
y = data.iloc[:,7:8].astype('float')
X.shape
train_stats = data.describe()
train_stats.pop('ERP')
train_stats = train_stats.transpose()
ytrain_stats = data['ERP'].describe()
print(ytrain_stats)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
X.head()
y.head()
```

```
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(X_train)
normed_test_data = norm(X_test)
normed_train_output = (y_train - ytrain_stats['mean']) / ytrain_stats['std']
print(normed_train_output)
print(normed_test_data)
normed_test_output = (y_test - ytrain_stats['mean']) / ytrain_stats['std']
print(normed_test_output)
```

```
model = tf.keras.Sequential()
```

```
model.add(Dense(6, activation='relu', input_shape=[7]))
model.add(Dense(4, activation='relu'))
model.add(Dense(1))
```

```
Optimizer = tf.keras.optimizers.SGD(lr=0.02)
model.compile(optimizer=Optimizer,loss='mse',metrics=['mae'])
model.summary() 
```

```
prediction = model.fit(normed_train_data, normed_train_output, epochs=100)
```

```
model.evaluate(X_test)
```
